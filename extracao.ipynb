Extração de texto completo, metadados e quantidade de páginas de um PDF usando a biblioteca PyMuPDF (fitz)

```python
import fitz  #PyMuPDF

#Caminho do pdf do qual será extraído o texto
PDF_PATH = "pdfs_teste/download.pdf"

#Abre o PDF
documento = fitz.open(PDF_PATH)

#Extrai os metadados do PDF para uma variável dicionário
metadados = documento.metadata
    
texto_completo = ""
    
#Itera sobre as páginas do PDF e extrai o texto de cada uma
for num_pagina in range(len(documento)):
    pagina = documento.load_page(num_pagina)
    
    #Adiciona o texto da página ao texto completo
    texto_completo += pagina.get_text()

print("Texto extraído do PDF: ", texto_completo)

print("\nMetadados do PDF: ")
#Itera sobre os metadados e imprime a chave e o valor de cada um
for chave, valor in metadados.items():
    print(f"{chave}: {valor}")
    
print("\nNúmero de páginas do PDF: ", len(documento))

#Fecha o documento pdf
documento.close()
```

Atualização de metadados usando PyMuPDF (fitz)

```python
#Caminho do pdf que será atualizado
PDF_PATH = "pdfs_teste/download.pdf"

#Abre o PDF
documento = fitz.open(PDF_PATH)

#Determina em uma variável dicionário os novos metadados que serão inseridos no PDF
novos_metadados = {
    "title": "Desastres naturais e coping coletivo",
    "author": "Fernanda Menna Barreto Krum, Denise Ruschel Bandeira",
    "subject": "Meio ambiente",
    "keywords": "Desastres naturais, Estilos de enfrentamento, Ambientes sociais"
}

#Atualiza os metadados do PDF
documento.set_metadata(novos_metadados)

#Caminho do novo PDF que será salvo
NEW_PDF_PATH = "pdfs_teste/download_atualizado.pdf"

#Salva o PDF atualizado
documento.save(NEW_PDF_PATH)

#Abre o PDF atualizado
documento_atualizado = fitz.open(NEW_PDF_PATH)

#Extrai os metadados do PDF atualizado para uma variável dicionário
metadados_atualizados = documento_atualizado.metadata

#Imprime os metadados atualizados para conferência
print("\nMetadados Atualizados: ")
for chave, valor in metadados_atualizados.items():
    print(f"{chave}: {valor}")
    
#Fecha ambos os documentos pdf
documento.close()
documento_atualizado.close()
```

Extração de texto completo, metadados e quantidade de páginas de um PDF usando a biblioteca PyPDF2

```python
pip install pypdf2 #Comando para instalar a biblioteca PyPDF2
```

```python
import PyPDF2

#Caminho do pdf do qual será extraído o texto
PDF_PATH = "pdfs_teste/download (1).pdf"

#Abre o PDF
with open(PDF_PATH, "rb") as arquivo_pdf:

    #Cria um objeto leitor do PDF
    leitor_pdf = PyPDF2.PdfReader(arquivo_pdf)
        
    texto_completo = ""

    #Itera sobre as páginas do PDF e extrai o texto de cada uma
    for pagina in range(len(leitor_pdf.pages)):
        pagina_pdf = leitor_pdf.pages[pagina]
        
        #Adiciona o texto da página ao texto completo
        texto_completo += pagina_pdf.extract_text()

#Abre o PDF novamente
with open(PDF_PATH, "rb") as arquivo_pdf:
    #Cria um objeto leitor do PDF
    leitor_pdf = PyPDF2.PdfReader(arquivo_pdf)
    #Extrai os metadados do PDF
    metadados = leitor_pdf.metadata
    #Determina o número de páginas do PDF
    numero_paginas = len(leitor_pdf.pages)
    

print("Texto extraído do PDF: ", texto_completo)

#Imprime os metadados do PDF
print("\nMetadados do PDF: ")
for chave, valor in metadados.items():
    print(f"{chave}: {valor}")

print("\nNúmero de páginas do PDF: ", numero_paginas)
```
